```python
import re
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
f = open("WhatsApp Chat with Hasnain Brother ğŸ¦¯.txt",'r',encoding='utf-8')
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:AM|PM)\s-\s[^:]+:'
messages = re.split(pattern, data)[1:]
len(messages)
```

```python
messages = re.split(pattern, data)[1:]

# remove leading commas, spaces, newlines
messages = [re.sub(r'^[,\s]+', '', m) for m in messages]

len(messages)
```

```python
messages
```

```python
dates = re.findall(pattern,data)
```

```python
df = pd.DataFrame({'user_message':messages, 'message_Date':dates})
#convert  into datetime.

df['message_Date'] = df['message_Date'].str.extract(
    r'(\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:AM|PM))'
)
```

```python
df['message_Date'] = pd.to_datetime(
    df['message_Date'],
    format='%m/%d/%y, %I:%M %p'
)
```

- Let seperate the message and user name.

```python
users = []
msgs = []

for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message, maxsplit=1)
    
    if len(entry) > 1:
        users.append(entry[1])
        msgs.append(entry[2])
    else:
        users.append('group_notification')
        msgs.append(entry[0])

df['users'] = users
df['messages'] = msgs
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['message_Date']
```

```python
df['year'] = df['message_Date'].dt.year
df['month'] = df['message_Date'].dt.month_name()
df['day'] = df['message_Date'].dt.day
df['hourse'] = df['message_Date'].dt.hour
df['minute'] = df['message_Date'].dt.minute
```
